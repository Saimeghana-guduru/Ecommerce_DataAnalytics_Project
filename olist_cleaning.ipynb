# **Orders Table**

```python
#data(we need to check that payment_seq=1 and pay_val=0 orders ids exist in orders table thats why imported this data)
df5=pd.read_csv('olist_orders_dataset.csv')
```

```python
#checking the 3 order ids with pay_val=0 and pay_seq =1 and whith no further orders exist in orders table
df5[df5['order_id'].isin([
    'c8c528189310eaa44a745b8d9d26908b',
    '4637ca194b6387e2d538dc89b124b0ee',
    '00b1cb0320190ca0daa2c88b35206009'
])]
#they are in orders table but they are cancelled so no need to remove them from payments table as they are valid order ids that are cancelled orders
```

```python
#Analysing
df5.head()
```

```python
#Null values
df5.isnull().sum()
#no prblm if we have null values in order_approved_at,order_delivered_carrier_date,order_delivered_customer_date..bcoz they are may be cancelled orders
```

```python
#duplicates
df5.duplicated().sum()
```

```python
#data types
df5.dtypes
```

```python
#size
df5.shape
```

```python
#changing the date cols from object to datetime
date_cols=[
    'order_purchase_timestamp',
    'order_approved_at',
    'order_delivered_carrier_date',
    'order_delivered_customer_date',
    'order_estimated_delivery_date'
]
df5[date_cols]=df5[date_cols].apply(pd.to_datetime)
```

```python
#checking datatypes again
df5.dtypes
```

```python
#checking chronological order of the dates
invalid1=df5[
    (df5['order_delivered_carrier_date'].notna()) &
    (df5['order_delivered_customer_date'].notna()) &
    (df5['order_delivered_carrier_date']>df5['order_delivered_customer_date'])
]

invalid2=df5[
    (df5['order_purchase_timestamp'].notna()) &
    (df5['order_approved_at'].notna()) &
    (df5['order_purchase_timestamp']>df5['order_approved_at'])
]

invalid3=df5[
    (df5['order_delivered_carrier_date'].notna()) &
    (df5['order_approved_at'].notna()) &
    (df5['order_approved_at']>df5['order_delivered_carrier_date'])
]

print("purchased > approved invalid orders = ",len(invalid2),
     "\napproved > carrier invalid orders = ",len(invalid3),
     "\ncarrier > delivery invalid orders = ",len(invalid1)
     )
#res = we can change the dates...may be they are wrongly entered bcoz they have to be get from diffrent sysytems like seller,courier,buyer ... and also Not deleting anything because they did not make any diff while calc yearly or monthly orders among 1,00,000 orders
```

```python
#Feature engineering(added delivery delay and status cols)

df5['delivery_delay_days']=(df5['order_delivered_customer_date']-df5['order_estimated_delivery_date']).dt.days
df5['delivery_status']=np.select(
    [
        df5['delivery_delay_days']<0,
        df5['delivery_delay_days']==0,
        df5['delivery_delay_days']>0
    ],
    [
        'Early',
        'On time',
        'Late'
    ],
    default=None
)
df5.head()
```
